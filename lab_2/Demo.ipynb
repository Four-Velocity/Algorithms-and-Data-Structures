#### Хеш-таблиця
[.data_structures.hash_map.py](https://github.com/Four-Velocity/Algorithms-and-Data-Structures/blob/master/lab_2/data_structures/hash_map.py) 

Класс CircleHashMap реалізує хеш-таблицю яка може містити у собі коло, ключем яких є периметр.
Колізії вирішуються методом подвійного хешування. Тобто кожного разу за умови коліїї вираховується додатковий хеш, незалежний від основного та додається до основного, таким чином вказуючи на нове мізце у хеш-таблиці.
Реальна довжина мапи на початку дорівнює 2<sup>1</sup>. З кожним запитом на збільшення таблиці ступінь двійки збільшується на 1.

__Методи__

| Ім'я | Опис |
|------|------|
|`__init__`|Ініціалізує усі необхідні для роботи таблиці атрибути| 
|`__len__`|Дозволяє отримувати кількість елементів у хеш-таблиці|
|`__getitem__`<br>`__setitem__`<br>`__delitem__`| Реалізують додавання об'єкту до таблиці, отримання значення за ключем, та вилучення з таблиці|
|`__iter__`|Ітерується по ключах хеш-таблиці|
|`__contains__`|Перевірка на знаходження об'єкту у таблиці|
|`__repr__`<br>`__str__`|Рядковий варіант таблиці. str - зрозумілий користувачу, repr - інтерпретатору|
|`__resize`|Змінює реальний розмір хеш-таблиці. Також перераховує значення хешу тіх елементів, що вже є в таблиці.|
|`__get_real_length`|Повертає реальний розмір хеш-таблиці з урахуванням всіх гепів|
|`__get_main_hash`|Повертає хеш об'єкту|
|`__get_fallback_hash`|Повертає додатковий хеш об'єкту для вирішення коллізій|
|`keys`|Повертає усі ключі з таблиці|
|`values`|Повертає усі значення з таблиці|
|`items`|Повертає усі пари ключ:значення з таблиці|
|`sort_delete`|Видаляє усі елементи що задовільняють критерію|

#### Виконання завдання

```python
from random import random, randint
from data_structures import CircleHashMap, Circle

# створимо хеш-таблицю...
hash_map = CircleHashMap()
# ... та заповнимо випадковими колами

for _ in range(25):
    x = round(randint(-10,10) + random(), 4)
    y = round(randint(-10,10) + random(), 4)
    r = round(randint(1,20) + random(), 4)
    circle = Circle(x,y,r)
    hash_map[circle.P] = circle
    
print(str(hash_map))
```

```python
# встановлюємо верхню межу площі та видаляємо всі елементи з площею нижче за вказану

area_ceil = 250
hash_map.sort_delete(lambda x : x.A < area_ceil)
```
